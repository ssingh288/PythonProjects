```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
```

```python
yt_df = pd.read_csv("howitsmade_season29")
```

```python
yt_df.head()
```

The index became "Unnamed: 0" column, I'll drop it.

```python
yt_df.drop(["Unnamed: 0"], axis=1, inplace=True)
```

```python
yt_df
```

```python
# Total number of views :

totalviews = yt_df.NB_OF_VIEWS.sum()
totalviews
```

```python
# Most viewed episode : 

mostviewed = yt_df[yt_df['NB_OF_VIEWS']==yt_df.NB_OF_VIEWS.max()]['SUBJECT']
mostviewed
```

```python
leastviewed = yt_df[yt_df['NB_OF_VIEWS']==yt_df.NB_OF_VIEWS.min()]['SUBJECT']
leastviewed
```

```python
yt_df
```

```python
# Calculating the Reaction % : out of all the views, what % of ppl liked or disliked.
yt_df["REACTION_PERC"] = round(((yt_df["LIKES"]+yt_df["DISLIKES"])/yt_df['NB_OF_VIEWS'])*100,2)
```

```python
yt_df[["SUBJECT","REACTION_PERC"]]
```

Art Glass Wall Sconces had least number of views, but people reacted the most to it.

```python
yt_df[["SUBJECT","REACTION_PERC"]].sort_values(by=['REACTION_PERC'],ascending=False).head(3)
```

Let's now look at the reactions themselves : love or hate ?

```python
yt_df["LIKE_PERC"]=round(yt_df["LIKES"]/(yt_df["LIKES"]+yt_df["DISLIKES"])*100,2)
yt_df["DISLIKE_PERC"]=round(yt_df["DISLIKES"]/(yt_df["LIKES"]+yt_df["DISLIKES"])*100,2)
```

```python
yt_df[["SUBJECT","LIKE_PERC","DISLIKE_PERC"]]
```

```python
yt_df[["SUBJECT","LIKE_PERC","DISLIKE_PERC"]].sort_values(by=['LIKE_PERC'],ascending=False).head(3)
```

```python
yt_df[["SUBJECT","LIKE_PERC","DISLIKE_PERC"]].sort_values(by=['DISLIKE_PERC'],ascending=False).head(3)
```
