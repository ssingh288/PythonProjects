```python
#16 caractères, 4 de chaque type

import random

#tous les caractères speciaux ne sont pas inclus, manque 58 à 64, 91 à 96, 123 à 126
def password ():
    
    pw = []
    
    for i in range(4):
        
        special = chr(random.randint(33,47))
        nbs = chr(random.randint(48,57))
        maj = chr(random.randint(65,90))
        lower = chr(random.randint(97,122))
        megalist = [special,nbs,maj,lower]
        
        for i in megalist :
            
            pw.append(i)
            
    random.shuffle(pw)
    
    pw = "".join(pw)
        
    return print("Your password is :",pw, " Please don't lose it.")

password()
```
